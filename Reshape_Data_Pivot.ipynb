```python
import pandas as pd
```

```python
# Sample input DataFrame

data = {'city'       : ['Jacksonville', 'Jacksonville', 'Jacksonville', 'Jacksonville', 'Jacksonville','ElPaso', 'ElPaso', 'ElPaso', 'ElPaso', 'ElPaso'],
        'month'      : ['January', 'February', 'March', 'April', 'May','January', 'February', 'March', 'April', 'May'],
        'temperature': [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]}

weather = pd.DataFrame(data)
weather
```

```python
# pivot the DataFrame by city

pivoted = weather.pivot(index='city', columns='month', values='temperature')
pivoted
```

```python
# pivot the DataFrame by month

pivoted = weather.pivot(index='month', columns='city', values='temperature').reset_index()
pivoted
```

```python
# pivot the DataFrame by month and sort it by order

month_order = ['January', 'February', 'March', 'April', 'May']
pivoted['month'] = pd.Categorical(pivoted['month'], categories=month_order, ordered=True)
pivoted = pivoted.sort_values('month')
pivoted
```
