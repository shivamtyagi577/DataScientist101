```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

**Defining Plot Color**

```python
x = range(1,10)
y = [1,6,5,0.5,2,4,5,7,4]

plt.bar(x,y)
```

**Let's change the Width of the Bar, Alignmet and Color**

```python
wide = [0.5, 0.5, 0.5, 0.9, 0.9, 0.9, 0.5, 0.5, 0.7]
color = ['salmon']

plt.title('Example of Customize Width, Color and Alignment')

plt.bar(x,y, width=wide, color=color, align ='center')
```
