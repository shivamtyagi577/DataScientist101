### Boolean Indexing

Comparsion Operators (<,>,=,<=,>=,)

And now you can see we have a DataFrame full of Boolean values where if the value is greater than 30, then the result is returned as true. And then if the value is less than 30, the result is returned as false


```python
mask = numbers_df > 30
mask
```

We can use this Boolean mask to retrieve a subset of data from the DataFrame. If we pass the Boolean mask in the DataFrame, it will return to us those values which were greater than 30. So we'll say numbers_df and then we just want to return the mask. And if we run this code block, we can see that only the values which are greater than 30 are returned. If the value is less than 30, then we get NAN, not a number. So if the values are less than 30, then we just get this NaN. So as you can see, this just returns all the values from the DataFrame that are greater than 30 as defined by our mask object. Boolean mask is also used to replace values in a DataFrame.


```python
numbers_df[mask]
```

```python
numbers_df[numbers_df> 30] = 0
numbers_df
```

# Working with Missing Data in Pandas

```python
data = {'name'   : ['Steve', 'john', 'richard', 'sarah', 'randy', 'michel', 'julie'],
        'age'    : [20,22,20,21,24,23,22],
        'gender' : ['Male','Male','Male','Female','Male','Male','Female'],
        'rank'   : [2,1,4,5,3,7,6]}

ranking_df = DataFrame(data)
ranking_df
```

```python
ranking_df.iloc[2:5,1] = np.nan
ranking_df.iloc[3:6,3] = np.nan
ranking_df.iloc[3,:] = np.nan
ranking_df
```

```python
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

#### To show you how to apply Boolean masking to show only the rows where there is a missing value in a specific column

```python
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
bool_series1 = ranking_df['age'].isnull()
bool_series1
```

```python
ranking_df[bool_series1]
```

### How to fill missing values in a Data Set

```python
ranking_df.fillna(0)
```

Let's fill missing values with the values that comes prior to the missing value within the data frame.

```python
ranking_df.fillna(method = 'pad')
```

let's fill missing values with the values that comes next to the missing value within the data frame.

```python
ranking_df.fillna(method = 'bfill')
```

```python
ranking_df.interpolate(method = 'linear')
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how = 'all')
```

```python
ranking_df.dropna(axis = 0)
```

```python
ranking_df.dropna(axis = 1)
```

### Concatenating Data and Transforming

```python
pd.concat([DF_obj, DF_obj_2], axis = 1)
```

```python
pd.concat([DF_obj, DF_obj_2], axis = 0)
```
