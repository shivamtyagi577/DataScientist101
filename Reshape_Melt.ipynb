```python
import pandas as pd
```

```python

# Sample input DataFrame
data = {'product'  : ['Umbrella', 'SleepingBag'],
        'quarter_1': [417, 800],
        'quarter_2': [224, 936],
        'quarter_3': [379, 93],
        'quarter_4': [611, 875]}

report = pd.DataFrame(data)

report
```

```python
# Reshape the DataFrame from wide to long format
reshaped = report.melt(id_vars='product',
                       value_vars=['quarter_1', 'quarter_2', 'quarter_3', 'quarter_4'],
                       var_name='quarter',
                       value_name='sales')

reshaped
```
